```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set a seed for reproducibility
np.random.seed(42)

# Create dummy data
data = np.random.normal(size=100)

# Plot a single violin plot
plt.figure(figsize=(8, 6))

# Create a violin plot with inner quartiles shown
sns.violinplot(data=data, inner="quart", bw=0.2)

# Get statistical values for annotation
median_value = np.median(data)
q25 = np.percentile(data, 25)
q75 = np.percentile(data, 75)
mean_value = np.mean(data)

# Align mean text with the dot but move it slightly up
plt.text(0.5, mean_value + 1, 'Mean ()', color='black', fontsize=12, verticalalignment='center')
plt.scatter(0, mean_value, color='black', edgecolor='black', s=100, label='Mean', zorder=5)

# Align median label with the median line
plt.text(0.5, median_value, 'Median (50th Percentile)', color='black', fontsize=12, verticalalignment='center')
plt.hlines(y=median_value, xmin=-0.25, xmax=0.25, color='black', linestyle='-', linewidth=1.5, zorder=5)

# Align 25th and 75th percentile labels with dashed lines
plt.text(0.5, q25, '25th Percentile', color='black', fontsize=12, verticalalignment='center')
plt.text(0.5, q75, '75th Percentile', color='black', fontsize=12, verticalalignment='center')
plt.hlines(y=[q25, q75], xmin=-0.25, xmax=0.25, color='black', linestyle='--', linewidth=1.5)

# Add a density indicator below the violin plot
plt.text(0, -3.3, "|------ density ------|", ha='center', fontsize=12, color='black')
#plt.plot([-0.15, 0.15], [-3, -3], color='gray', lw=3)

# Title
plt.title('Violin Plot Legend', fontsize=16)

# Hide the x-axis and make the plot look cleaner
plt.xticks([])
plt.yticks([])
plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```
